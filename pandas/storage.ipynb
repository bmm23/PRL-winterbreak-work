```python
import pandas as pd
import numpy as np

```

```python
url = "https://raw.githubusercontent.com/fivethirtyeight/nfl-elo-game/"
url = url + "3488b7d0b46c5f6583679bc40fb3a42d729abd39/data/nfl_games.csv"

nfl = pd.read_csv(url)
nfl.head()
print(f'Shape: {nfl.shape}, \nColumns: {nfl.columns}')

```

```python
nfl.to_csv()
```

# Analysis 

```python
nfl['totalpts'] = nfl['score1'] + nfl['score2']
nfl['playoffpts'] = nfl[nfl['playoff']==1]['score1'] + nfl[nfl['playoff']==1]['score2']
meanPts = nfl['totalpts'].mean()
print(f'Mean Points: {meanPts:.2f}')
meanPlayoff = nfl[nfl['playoff'] == 1]['playoffpts'].mean()
print(f'Mean Playoff Pts: {meanPlayoff:.2f}')

upset1 = nfl[nfl['elo1'] > nfl['elo2']]
realUpset1 = upset1[upset1['score1'] < upset1['score2']]

upset2 = nfl[nfl['elo1'] < nfl['elo2']]
realUpset2 = upset2[upset2['score1'] > upset2['score2']]

realUpset = pd.concat([realUpset1, realUpset2],ignore_index=True)

print(f'Upset Ratio: { len(realUpset)/ len(nfl):.2f}')
```
